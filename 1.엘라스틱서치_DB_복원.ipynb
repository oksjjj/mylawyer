````python
import json
from elasticsearch import Elasticsearch

# ===== ES ì—°ê²° =====
es_client = Elasticsearch(
    "https://localhost:9200",
    basic_auth=("elastic", "changeme"),
    verify_certs=False,
    ssl_show_warn=False
)

# ë°±ì—… íŒŒì¼ ê²½ë¡œ
file_path = "cases_full_backup.json"
new_index_name = "cases"   # ë³µì› ëŒ€ìƒ ì¸ë±ìŠ¤ ì´ë¦„

# ===== JSON ë¡œë“œ =====
with open(file_path, "r", encoding="utf-8") as f:
    docs = json.load(f)

print(f"ì´ ë¬¸ì„œ ìˆ˜: {len(docs)}")

batch_size = 1000

for i in range(0, len(docs), batch_size):
    batch = docs[i:i+batch_size]
    actions = []

    print(f"ì—…ë¡œë“œ ì¤‘: {i} ~ {i + len(batch) - 1} (ì´ {len(batch)})")

    for d in batch:
        # ì›ë˜ _id ìœ ì§€
        _id = d.get("_id")

        # ì›ë˜ ì¸ë±ìŠ¤ ê·¸ëŒ€ë¡œ ë³µêµ¬í• ì§€, ìƒˆ ì¸ë±ìŠ¤ë¡œ ë„£ì„ì§€ ì„ íƒ
        index_target = new_index_name or d["_index"]

        actions.append({
            "index": {
                "_index": index_target,
                "_id": _id
            }
        })

        # ES bulkëŠ” _source ë‚´ë¶€ë§Œ ë„£ì–´ì•¼ í•¨
        actions.append(d["_source"])

    es_client.bulk(body=actions, refresh=True)

print("ğŸ”¥ ë³µì› ì™„ë£Œ!")
````
