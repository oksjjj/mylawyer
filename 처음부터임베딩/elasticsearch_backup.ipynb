```python
from elasticsearch import Elasticsearch
import json

# ===== ES 연결 =====
es_client = Elasticsearch(
    "https://localhost:9200",
    basic_auth=("elastic", "changeme"),
    verify_certs=False,
    ssl_show_warn=False
)

index_name = "cases"
output_file = "cases_full_backup.json"

# ===== scroll 검색 시작 =====
scroll = "2m"
page = es_client.search(
    index=index_name,
    body={"query": {"match_all": {}}},
    scroll=scroll,
    size=1000,    # 한번에 가져올 개수
    _source=True  # vector 포함 전체 소스 가져옴
)

all_docs = []
scroll_id = page["_scroll_id"]
hits = page["hits"]["hits"]

print("초기 로드:", len(hits))

while hits:
    for h in hits:
        # ES 백업 시 필요한 필드만 추출 (_source 포함)
        all_docs.append({
            "_id": h.get("_id"),
            "_index": h.get("_index"),
            "_score": h.get("_score"),
            "_source": h.get("_source"),  # text + metadata + vector 모두 포함
        })

    page = es_client.scroll(scroll_id=scroll_id, scroll=scroll)
    scroll_id = page.get("_scroll_id")
    hits = page["hits"]["hits"]
    print("로드 중, 현재 총:", len(all_docs))

# ===== JSON 파일로 저장 =====
with open(output_file, "w", encoding="utf-8") as f:
    json.dump(all_docs, f, ensure_ascii=False, indent=2)

print("백업 완료! 총 문서 수:", len(all_docs))
print("파일 저장:", output_file)
```
